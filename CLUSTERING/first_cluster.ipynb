```python
import pandas as pd
import numpy as np
```

### CLUSTERING FOR QUEENS

```python
data = pd.read_csv('queens.csv')
data
```

```python
del data[data.columns[0]]
```

```python
data
```

```python
data.FELONY = data.FELONY / data.population
data.MISDEMEANOR = data.MISDEMEANOR / data.population
data.VIOLATION = data.VIOLATION / data.population
data
```

```python
data = data[['community','FELONY','MISDEMEANOR','VIOLATION']]
```

```python
data
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
score = {}
x = data.iloc[:,1:]
for i in range(2,5):
    km = KMeans(random_state=999,n_clusters=i)
    res=km.fit(x)
    labels = res.labels_
    score[i] = silhouette_score(x, labels)
    print("For n_clusters = {},".format(i)+" the silhouette_score is : {}".format(score[i]))
print('The choosen number of clusters is {}'.format([n for n in score.keys() if score[n]==max(score.values())][0]))

km = KMeans(random_state=999,n_clusters=[n for n in score.keys() if score[n]==max(score.values())][0])
res=km.fit(x)
```

```python
data['labels'] = res.labels_
data
```

```python
 data.to_csv('queens_cluster.csv')
```
